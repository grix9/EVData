Electric Vehicle Data Washinton USA, 2020-2025

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

evdf = pd.read_csv('Electric_Vehicle_population_data.csv')
print('Total Count:')
print(evdf['VIN (1-10)'].count())
print(evdf.head(3))
```

```python
# Top 5 Most Popular EV Brands Pie Chart, Top 5 Most Popular Models

top_evs_brands = evdf['Make'].value_counts()
evdf['Make'].value_counts().head(5).plot.pie()
plt.title('Top 5 Most Popular EV Brands')
plt.axis("off")

top_models = evdf['Model'].value_counts()
top_model_text = top_models.head(5).to_string()
new_text = top_model_text.replace('Model', 'Top 5 Models:')
print(new_text)
```

```python
# MSRP Statistics
# Average MSRP, Brand with Highest MSRP, MSRP and Range Correlation Scatter Plot, Top Range, Average Range

#Remove zeros from MSRP
msrp_zeros = evdf['Base MSRP'].replace(0, np.nan)
msrp_zeros.dropna(inplace=True)

print('Average MSRP: \n' + str(msrp_zeros.mean().round(2)))


print('Brand with Highest MSRP:')
print(evdf[['Base MSRP', 'Make']].max())
print('\n')

#top range
top_range = evdf[['Electric Range']].max()
print('Top Electric Range: \n' + str(top_range))
#average range
range_zeros = evdf['Electric Range'].replace(0, np.nan)
range_zeros.dropna(inplace=True)
average_range = range_zeros.mean()
print('Average Electric Range: \n' + str(average_range.round(2)))

plt.scatter(x=evdf['Electric Range'], y=evdf['Base MSRP'])
plt.show()
plt.close()

sns.histplot(x = msrp_zeros, data = evdf,)
```

```python
# Most Popular Year, Average Year, Post Code with Most EVs

most_popular_year = evdf['Model Year'].value_counts().idxmax()
print("Most Popular Model Year: " + str(most_popular_year))

average_year = evdf['Model Year'].mean().round()
print('Average Model Year: ' + str(int(average_year)))

city_most_evs = evdf['City'].value_counts().idxmax()
print('City with Most EVs: ' + str(city_most_evs))

post_code_most_evs = evdf['Postal Code'].value_counts().idxmax()
print('Postal Code with Most EVs: ' + str(int(post_code_most_evs)))
```
